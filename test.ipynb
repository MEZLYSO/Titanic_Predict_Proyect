```python
import pandas as pd
import numpy as np
```

```python
# Cargamos los datos de entrenamiento para el modelo
train_data= pd.read_csv("train.csv")
train_data
```

```python
# Cargamos los datos de prueba para el modelo
test_data= pd.read_csv("test.csv")
test_data
```

```python
# Ahora rellenamos los valores faltantes en la columna 'Age' con la mediana de la edad
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
```

```python
# Como siguiente rellenamos los valores faltantes en la columna 'Embarked' con la moda y en 'Fare' con la mediana
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
# En este paso convertimos las variables categóricas en numéricas utilizando Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in ['Sex', 'Embarked']:
    train_data[col] = le.fit_transform(train_data[col])
    test_data[col] = le.fit_transform(test_data[col])
    
```

```python
# Definimos las caracteristicas que deseamos considerar 
# Ademas de el objetivo que deseamos encontrar que es que sobrevivio
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
target = 'Survived'
```

```python
# Ahora definimos nuestras caracteristicas y nuestro objetivo
X = train_data[features]
Y = train_data[target]
```

```python
# Ahora dividimos nuestros datos en entrenamiento y validacion 
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Ahora lo que tenemos que hacer es crear nuestro modelo y definimos las iteraciones y que solo va a tomar los datos de entranamiento
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=250)
model.fit(X_train, y_train)
```

```python
# Ahora importamos el acurracy_store para ver la precision del modelo
from sklearn.metrics import accuracy_score
# Tomamos los datos de validacion e intetamos predecir la validacion
y_pred = model.predict(X_val)
print("Validation Accuracy:", accuracy_score(y_val, y_pred))
```

```python
# Prueba final con los datos 
predictions = model.predict(test_data[features])
test_y = test_data[target]
pred_y = predictions
print("Validation Accuracy:", accuracy_score(test_y, pred_y))
```
